```python
import json
data = json.load(open('Taxcomputation.json'))
#data = data['dependencies']

extracted_data = []
for d in data:
    d_add = d['dependencies']['add']
    d_drop = d['dependencies']['drop']
        
    transformText = d['transformText']
    outputName = d['outputName']

    #print("add len: {}, drop len: {}".format(len(d_add), len(d_drop)))
    for item in d_add:
        if item['from']:
            ed = {}
            ed['transformText'] = transformText
            ed['outputName'] = outputName
            ed['dependencies.add.name'] = item['name']
            for from_item in item['from']:
                ed['dependencies.add.from.col'] = from_item['col'][0] if item['from'][0]['col'] else ''
                ed['dependencies.add.from.table'] = from_item['table']
                ed['dependencies.add.from.type'] = from_item['type']
                if d_drop:
                    ed['dependencies.drop.name'] = d_drop[0]['name']
                    ed['dependencies.drop.type'] = d_drop[0]['type']
                    ed['dependencies.drop.table'] = d_drop[0]['table']
                else:
                    ed['dependencies.drop.name'] = ''
                    ed['dependencies.drop.type'] = ''
                    ed['dependencies.drop.table'] = ''
                    
                extracted_data.append(ed)
```

```python
import pandas as pd
df = pd.DataFrame(extracted_data)
df.head()
```
